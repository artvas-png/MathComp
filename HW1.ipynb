```python
import time as tm

Gl_N=101

def mat_np(N):
  sum=0
  for i in range(0, 10):
    A1=np.random.randint(-10, 10, (N,N))
    A2=np.random.randint(-10, 10, (N,N))
    T1=tm.time()
    A=np.dot(A1, A2)
    T2=tm.time()
    sum+=(T2-T1)*100
  return sum

def mat_bh(N):
  for l in range(0, 10):
    A1=np.random.randint(-10, 10, (N,N))
    A2=np.random.randint(-10, 10, (N,N))
    D=np.zeros((N, N))
    sum=0
    T1=tm.time()
    for i in range(0, N):
      for j in range(0, N):
        for k in range(0, N):
         D[i][j]+=A1[i][k]*A2[k][j]
    T2=tm.time()
    sum+=(T2-T1)*100
  return sum

n=range(1,Gl_N)
T_np=[]
T_bh=[]

for N in range (1, Gl_N):
  T_np.append(mat_np(N))
for N in range (1, Gl_N):
  T_bh.append(mat_bh(N))


plt.figure(figsize=(10, 10))
plt.ylabel("Размер матрицы N", fontsize=16)
plt.xlabel(" Время перемножения t, мс", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.plot(n, T_np, label='Numpy product')
plt.plot(n, T_bh, label='Standart product')
plt.legend(loc='best', fontsize='x-large')
plt.grid(True)
plt.show()
```
