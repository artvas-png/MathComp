```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x=sns.load_dataset('anscombe')

x1, y1=x.values[0:11, 1:2], x.values[0:11, 2:3]
x2, y2=x.values[11:22, 1:2], x.values[11:22, 2:3]
x3, y3=x.values[22:33, 1:2], x.values[22:33, 2:3]
x4, y4=x.values[33:44, 1:2], x.values[33:44, 2:3]

print(np.cov(x1.reshape(1, 11).astype(float), y1.reshape(1, 11).astype(float)))
print(np.cov(x2.reshape(1, 11).astype(float), y2.reshape(1, 11).astype(float)))
print(np.cov(x3.reshape(1, 11).astype(float), y3.reshape(1, 11).astype(float)))
print(np.cov(x4.reshape(1, 11).astype(float), y4.reshape(1, 11).astype(float)))

vx1, vy1=np.var(x1), round(np.var(y1), 3)
vx2, vy2=np.var(x2), round(np.var(y2), 3)
vx3, vy3=np.var(x3), round(np.var(y3), 3)
vx4, vy4=np.var(x4), round(np.var(y4), 3)

plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.scatter(x1, y1)
plt.title('Dataset I'+r'$\sigma^2_{x1}=$'+str(vx1)+', '
+ r'$\sigma^2_{y1}=$'+str(vy1))
plt.grid(True)


plt.subplot(222)
plt.scatter(x2, y2)
plt.title('Dataset II, '+r'$\sigma^2_{x2}=$'+str(vx2)+', '
+ r'$\sigma^2_{y2}=$'+str(vy2))
plt.grid(True)

plt.subplot(223)
plt.scatter(x3, y3)
plt.title('Dataset III, '+r'$\sigma^2_{x3}=$'+str(vx3)+', '
+ r'$\sigma^2_{y3}=$'+str(vy3))
plt.grid(True)

plt.subplot(224)
plt.scatter(x4, y4)
plt.title('Dataset IV, '+r'$\sigma^2_{x4}=$'+str(vx4)+', '
+ r'$\sigma^2_{y4}=$'+str(vy4))
plt.grid(True)
```

```python


x=((np.random.random((1, 10))-5)*10)
y=((np.random.random((1, 10))-5)*10)



w1 = np.cov(np.vstack((x, y)))[0, 1]/ np.var(x)
w0 = np.mean(y) - w1 * np.mean(x)

t=np.arange(x.min(), x.max(), 0.01)

plt.figure(figsize=(7, 7))
plt.plot(t, w1 * t + w0)
plt.scatter(x, y)
plt.grid(True)
plt.show()

print(np.corrcoef(x, y))
```
